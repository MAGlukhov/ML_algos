# Одномероная структура данных Pandas - series

```python
data = [1,2,3,4,5]
series = pd.Series(data)
print(series)
```

# Двумерная структура данных - DataFrame

```python
data = {
        'Имя': ['Алексей', 'Евгения','Григорий'],
        'Возраст': [25, 35,22],
        'Город': ['Барнаул', 'Москва','Воронеж']
}

df = pd.DataFrame(data)
print(df)
```
