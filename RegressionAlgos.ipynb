# Простой подход линейной регрессии

```python
random.seed(1)
k = random.randint(0,100) / 100
random.seed(2)
b = random.randint(0,100) / 100
n1, n2 = 0.0001, 0.0001
epoch = 1000
```

```python
print(k,b, sep=' ')
```

```python
y1 = list()
for i in range(len(x)):
    y1.append(x[i]*k + b)
```

```python
print(y1)
print(y)
```

```python
plt.grid()
plt.plot(x,y1, 'ro')
plt.plot(x,y, 'bo')
```

```python
for j in range(epoch):
    for i in range(len(x)):
        if y[i] - y1[i] > 0 and x[i] > 0:
            k+=n1
            b+=n2
        if y[i] - y1[i] > 0 and x[i] < 0:
            k-=n1
            b+=n2
        if y[i] - y1[i] < 0 and x[i] > 0:
            k-=n1
            b-=n2
        if y[i] - y1[i] < 0 and x[i] > 0:
            k+=n1
            b-=n2
        y1[i] = k * x[i] + b
```

```python
print(k,b, sep=' ')
```

```python
plt.grid()
plt.plot(x,y1, 'ro')
plt.plot(x,y, 'bo')
```
